```python
import os
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PartitionManagement") \
    .master("local[*]") \
    .getOrCreate()
```

```python
df = spark.range(0, 5000000)
initial_partitions = df.rdd.getNumPartitions()
print(f"Initial number of partitions: {initial_partitions}")
print(f"Total record count: {df.count():,}")
```

```python
repartitioned_df = df.repartition(12)

new_partitions = repartitioned_df.rdd.getNumPartitions()
print(f"Number of partitions after repartition(): {new_partitions}")
```

```python
coalesced_df = repartitioned_df.coalesce(3)

final_partitions = coalesced_df.rdd.getNumPartitions()
print(f"Number of partitions after coalesce(): {final_partitions}")
```
